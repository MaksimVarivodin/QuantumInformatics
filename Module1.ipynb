## 1. Дані, повідомлення, інформація: в чому суть цих понять і співвідношення між ними?

- **Дані** — це необроблені факти й цифри, які самі по собі не мають сенсу.
- **Повідомлення** — це структуровані дані, які можуть передаватися від одного джерела до іншого. Вони можуть містити інформацію.
- **Інформація** — це дані, оброблені таким чином, що вони мають сенс і можуть використовуватися для прийняття рішень.

**Співвідношення**: дані — це сировина, з якої формуються повідомлення, а повідомлення, що містять оброблені дані, стають інформацією.

## 2. Міра інформації за Хартлі та Шенноном

- **Міра інформації за Хартлі**:

 $$I = \log_b(N)$$
  
  де $N$ — кількість можливих повідомлень, а $b$ — основа логарифма (зазвичай 2, що відповідає бітам).

- **Міра інформації за Шенноном**:

 $$
  I = -\sum_{i=1}^{n} p_i \log_b(p_i)
 $$

  де $p_i$ — ймовірність отримання $i$ -го повідомлення.

**Обґрунтування логарифмічної функції**: логарифм показує, як швидко зростає кількість інформації зі збільшенням кількості можливих повідомлень, що робить його зручним для вимірювання інформації.

## 3. Аналіз співвідношення між різними одиницями величини інформації

Для аналізу наведемо формули:

- **Біт**:

 $$
  I = -\log_2(P)
 $$

- **Нат**:

 $$
  I = -\ln(P)
 $$

- **Дит (децит)**:

 $$
  I = -\log_{10}(P)
 $$
**Графіки $\log(P)$ залежно від $P$ можна побудувати за допомогою Python:**




```python

import numpy as np
import matplotlib.pyplot as plt

P = np.linspace(0.01, 1, 100)
log2_P = -np.log2(P)
ln_P = -np.log(P)
log10_P = -np.log10(P)

plt.figure(figsize=(10, 6))
plt.plot(P, log2_P, label='-log₂(P)', color='b')
plt.plot(P, ln_P, label='-ln(P)', color='g')
plt.plot(P, log10_P, label='-log₁₀(P)', color='r')
plt.title('Графіки різних одиниць інформації')
plt.xlabel('P')
plt.ylabel('Інформація')
plt.legend()
plt.grid()
plt.show()
```

#### Перехід між одиницями

Коефіцієнти переходу між бітами, натами та дитами:

- З бітів у нати:

 $$
  1 \text{ біт} = \log_2(e) \approx 1.4427 \text{ нат}
 $$

- З бітів у дити:

 $$
  1 \text{ біт} = \log_{10}(2) \approx 0.3010 \text{ дит}
 $$

**Пояснення отриманих результатів**: ці коефіцієнти показують, як перевести кількість інформації з одних одиниць в інші, використовуючи властивості логарифмів.

## 4. Приклади залежності кількості інформації

- **Для міри Хартлі**: кількість інформації зростає зі збільшенням числа знаків алфавіту. Наприклад, у двійковій системі (2 знаки) інформація на символ менша, ніж у шістнадцятковій (16 знаків).

- **Для міри Шеннона**: якщо ймовірність отримання певного повідомлення менша, то інформація, яку воно несе, більша. Наприклад, рідкісні слова в мові несуть більше інформації, ніж часто вживані.

## 5. Взаємозв'язок між мірами Хартлі та Шеннона

- **Міра Хартлі** є окремим випадком міри Шеннона для рівноймовірних подій:

 $$
  I = \log_b(N) = -\sum_{i=1}^{N} \frac{1}{N} \log_b\left(\frac{1}{N}\right)
 $$

- **Алгоритмічне визначення Колмогорова**: інформація вимірюється довжиною найкоротшого алгоритму (програми), який може згенерувати дані.

## 6. Взаємозв'язок інформації з іншими фізичними величинами

- **Принцип Ландауера**: стирання одного біта інформації вимагає мінімальної енергії, визначеної як:

 $$
  E = k T \ln 2
 $$

  де $k$ — стала Больцмана, $T$ — температура в Кельвінах.

- **Зв'язок з ентропією**: інформація та ентропія пов'язані обернено пропорційно; збільшення інформації зменшує ентропію системи.

- **Непозиційні системи**: значення цифр не залежить від їхньої позиції (римські цифри).

## 9. Інформативна ємність різних систем числення

**Вказівка**: якщо за основу системи числення взяти число $x$, то у системі з $n$ числами отримаємо $\frac{n}{x}$ розрядів. Число цифр, які при цьому можна записати, дорівнює $x^{\frac{n}{x}}$.

**Функція**:

$$
y(x) = n \left( x^{\frac{n}{x}} \right) = n x^{x}
$$

**Для встановлення інформативної ємності потрібно знайти максимум функції $y(x)$ :**

$$
\frac{dy}{dx} = n x^{x} (\ln x + 1) = 0
$$

**Доведення співвідношення:**

1. Прологарифмуємо функцію:

  $$
   \ln y = \ln (n x^x) = \ln n + x \ln x
  $$

2. Знайдемо похідну:

  $$
   \frac{d (\ln y)}{dx} = \frac{1}{y} \frac{dy}{dx} = \frac{d}{dx} [\ln n + x \ln x] = \ln x + 1
  $$

3. Тоді похідна функції $y(x)$ :

  $$
   \frac{dy}{dx} = y (\ln x + 1)
  $$

4. Прирівнюємо до нуля:

  $$
   y (\ln x + 1) = 0 \implies \ln x + 1 = 0 \implies x = e^{-1}
  $$

Оскільки основа системи числення $x$ має бути більше 1, максимума функція не досягає в цій точці в області визначення, тому вона монотонно зростає для $x > 1$ .

**Побудова графіків** за допомогою Python:



```python
import numpy as np
import matplotlib.pyplot as plt

def y(x, n):
    return n * x ** x

x_values = np.linspace(1, 5, 400)
plt.figure(figsize=(10, 6))

for n in [2, 8, 10, 16]:
    y_values = y(x_values, n)
    plt.plot(x_values, y_values, label=f'n={n}')

plt.title('Інформативна ємність різних систем числення')
plt.xlabel('x (основа системи числення)')
plt.ylabel('y(x)')
plt.legend()
plt.grid()
plt.show()
```

## 10. Інформація, що переноситься однією літерою українського алфавіту

**Для рівномірного розподілу** ($p_i = \frac{1}{32}$ ):

$$
I = -\sum_{i=1}^{32} \frac{1}{32} \log_2\left(\frac{1}{32}\right) = \log_2(32) = 5 \text{ біт}
$$

**Для нерівномірного розподілу** (з використанням таблиці частот):



```python
import numpy as np

frequencies = {
    "SPACE": 0.134, "о": 0.082, "н": 0.070, "а": 0.070, "и": 0.056,
    "т": 0.051, "в": 0.046, "е": 0.043, "р": 0.038, "і": 0.037,
    "с": 0.036, "к": 0.036, "м": 0.033, "л": 0.028, "д": 0.028,
    "у": 0.027, "п": 0.025, "я": 0.021, "з": 0.019, "ь": 0.015,
    "г": 0.013, "ч": 0.011, "б": 0.010, "х": 0.010, "ц": 0.009,
    "ю": 0.009, "й": 0.007, "ї": 0.006, "є": 0.006, "ф": 0.005,
    "ш": 0.005, "щ": 0.003, "ґ": 0.000
}

I = -sum(p * np.log2(p) for p in frequencies.values() if p > 0)
print(f"Середня кількість інформації на символ: {I:.4f} біт")
```

**Пояснення отриманих результатів**: середня інформація на символ менша за 5 біт через нерівномірність розподілу частот літер. Частіші літери несуть менше інформації, тоді як рідкісні — більше. Це демонструє, як ймовірність появи символу впливає на кількість інформації.
